### hw1.A: (10 points) practice of conditional selection using if, if-else, or if-elif-else.
Your program will ask user to input a score between 0 to 100 inclusive.  
It will output letter grade A, B, C, D, F on the screen accordingly.
 
 - write a short description for this homework
 - use keyboard input for score, 
 - use if-elif-else to convert score [0-100] to A, B, C, D, F
 - output should be A, B, , F. 
 
 


# sample run:


```python
# your code here 
'''
This program will take user input, convert the input to an integer, then check to see what grade the integer corresponds to. If the input is out of the range 0-100 the program will output an error message
'''
# get the score from a user's keyboard input and convert the str input to int type
usr_input = input('What is your score? (0-100)')
score = int(usr_input)                                      # convert user input into integer

if 0 <= score < 60:                                         # check what range the score falls into
    print('Your grade is: F')
elif score < 70:
    print('Your grade is: D')
elif score < 80: 
    print("your grade is: C")
elif score < 90:
    print('Your grade is: B')
elif score <= 100:
    print('Your grade is: A')
else:
    print('Input out of range.')                            # if input < 0 or > 100, print invalid input














```
