## B2. Missing Values (10 points)

```python
#  use comment line to describe what the each code line does
 
flights.isnull()
#returns a boolean series indicating whether each value is True (null) or False (not null) 
flights.isna()
#returns a boolean series indicating whether each value is True (N/A) or False (not N/A) 


flights.notna()
# exact opposite of .isna(), returns boolean series whether each value is Trie (not N/A) or False (N/A)
```

```python
# df_miss:  create a datafram such that each row has at least one missing value, display the first 5 rows

df_miss = pd.DataFrame(np.random.randint(0,100,size=(100, 4)), columns=list('ABCD'))

for i in range(df_miss.shape[0]):
    df_miss.iloc[i, np.random.randint(0,4)] = np.nan
print(df_miss.head())
# print out the number of rows in df_miss:
print(df_miss.shape[0])

```

```python
# Use the dataframe flights:
# Calculate how many rows were dropped if you drop any row which has at least one missing value
difference = flights.shape[0] - flights.dropna().shape[0]
print('Number of rows with at least one missing value:', difference)





```

```python
# Chech the difference between the following dropna options: 

# USE COMMENT TO MAKE NOTE


flights.dropna(axis = 0)
# drops all ROWS with at least one missing value

flights.dropna(axis = 1)
# drops all COLOUMNS with at least one missing value
```

```python
# Chech the difference between the following dropna options: WRITE Comments


flights.dropna(how ='any').shape
# This option drops less rows than the one below, but the number of columns stay the same

flights.dropna(how ='all').shape 
# this only drops a row if all the values are NA
```

```python
# flight1: using boolean index
# Using dataframe flights to create flight1 by filtering out all the rows where "dep_delay" value is missing, 
# i.e. keep only the row where "dep_delay" value is not a missing value
flight1 = flights.dropna(subset='dep_delay')
flight1



# print out how many rows are kept.
print(flight1.shape[0])


```

```python
# flight2: either/both

# create flight2 which has "dep_delay" and "arr_delay"columns only, print out the number of rows
 
flight2 = flights[['dep_delay', 'arr_delay']]
print(flight2.head())
print(flight2.shape[0])
    
# for flight2, find and print out how many rows such that there is at least a missing value 
print(flight2.shape[0] - flight2.dropna().shape[0])



# for flight2, find and print out how many rows such that both are missing
print(flight2.shape[0] - flight2.dropna(how='all').shape[0])
    
    
```
